```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('100CANON/IMG_9371.JPG')

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.scatter(5130,3400,c='w')
```

```python
plt.figure(figsize=(20,10))
pts1 = np.float32([[314,246],[5020,123],[143,3422],[5130,3400]])
pts2 = np.float32([[0,0],[3000,0],[0,3000*2//3],[3000,3000*2//3]])

rows,cols,ch = img.shape


M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(3000,3000*2//3))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
plt.imsave('this.jpg',cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
```
